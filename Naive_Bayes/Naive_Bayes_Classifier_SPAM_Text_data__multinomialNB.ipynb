```python
bbc["spam"]=bbc["v1"].apply(lambda x:1 if x=="spam" else 0)
bbc.head()
```

#spliting data in X,y

```python
X = bbc[['v2']] # Indepentent feature
y = bbc['v1'] #Dependent feature
```

```python
y.head(100)
```

```python
X.tail()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.25,
                                                    random_state=42,
                                                    stratify=y)
```

```python
# Import CountVectorizer.
from sklearn.feature_extraction.text import CountVectorizer

# Instantiate our CountVectorizer.
cvec = CountVectorizer(max_features=500 , stop_words= 'english')
```

```python
# Fit our CountVectorizer on the training data and transform training data.
X_train_cvec = pd.DataFrame(cvec.fit_transform(X_train['v2']).todense(),
                            columns=cvec.get_feature_names_out())
```

```python
# Transform our testing data with the already-fit CountVectorizer.
X_test_cvec = pd.DataFrame(cvec.transform(X_test['v2']).todense(),
                            columns=cvec.get_feature_names_out())
```

```python
X_train_cvec.shape
```

```python
X_train_cvec.head()
```

```python
# Import our model!

from sklearn.naive_bayes import MultinomialNB
# Instantiate our model!

nb = MultinomialNB()
```

```python
# Fit our model!

model = nb.fit(X_train_cvec,y_train)
```

```python
# Generate our predictions!

predictions = model.predict(X_test_cvec)
```

```python
# Score our model on the training set.

model.score(X_train_cvec,y_train)
```

```python
# Score our model on the testing set.

model.score(X_test_cvec,y_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,predictions)
```

```python
tn, fp, fn, tp = confusion_matrix(y_test,predictions).ravel()
```

```python
print("True Negative : %s "%tn)
print("False Positive : %s "%fp)
print("False Negative : %s "%fn)
print("True Positive : %s "%tp)
```
