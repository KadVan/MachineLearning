```python
# Convert all columns to lowercase and replace spaces in column names.#### as per sir
tip.columns = tip.columns.str.lower().str.replace(' ', '_')
tip.head()
```

```python
#### The difference is not good. it is either overfit or underfit,so using lasso and ridge for clear it.##as per notebook---first we do linear regression with polynomial features.(regular regression)
```

```python
# Create X and y.
X = tip.drop('tip', axis=1)
y = tip['tip']
```

```python
# Instantiate our PolynomialFeatures object to create all two-way terms.
# Write the code
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=2, interaction_only=False, include_bias=False)
# Fit and transform our X data.
X_overfit=poly.fit_transform(X)
```

```python
poly.get_feature_names(X.columns)
```

```python
# Check out the dimensions of X_overfit.
X_overfit.shape
```

```python
###Let's split our data up into training and testing sets. Why do we split our data into training and testing sets
```

```python
# Import train_test_split.
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
# Create train/test splits.      ##### to show overfit modl test size is 70% taken by teacher####
X_train, X_test, y_train, y_test = train_test_split(
    X_overfit,
    y,
    test_size=0.7,
    random_state=42
)
```

```python
# Scale our data.
# Relabeling scaled data as "Z" is common.
# write the code

sc = StandardScaler()
Z_train= sc.fit_transform(X_train)
Z_test= sc.transform(X_test)
```

```python
print(f'Z_train shape is: {Z_train.shape}')
print(f'y_train shape is: {y_train.shape}')
print(f'Z_test shape is: {Z_test.shape}')
print(f'y_test shape is: {y_test.shape}')
```

```python
# Import the appropriate library and fit our OLS model.#####    OLS is my linearregression#####

from sklearn.linear_model import LinearRegression
```

```python
ols = LinearRegression()
ols.fit(Z_train,y_train)
```

```python
# How does the model score on the training and test data?######  THIS IS  OVERFIT MODEL,SO DO HERE somekind of exercise.
print(ols.score(Z_train,y_train))
print(ols.score(Z_test,y_test))
```

```python
ols.coef_
```

```python
#### DO HERE implementation
```

```python
##Ridge implementation######
```

```python
# Ridge regressor lives here:
from sklearn.linear_model import Ridge
```

```python
# Instantiate.
ridge_model = Ridge(alpha=10)

# Fit.
ridge_model.fit(Z_train,y_train)

# Evaluate model using R2.
print(ridge_model.score(Z_train,y_train))
print(ridge_model.score(Z_test,y_test))
```

```python
###do iteration because sill now our model is overfit.so we do iteration with help of thisequation---
from sklearn.linear_model import RidgeCV
```

```python
# Set up a list of ridge alphas to check.
# np.logspace generates 100 values equally between 0 and 5,
# then converts them to alphas between 10^0 and 10^5.
r_alpha = np.logspace(0,5,100)

# Cross-validate over our list of ridge alphas.
ridge_cv = RidgeCV(alphas = r_alpha, scoring= 'r2', cv = 5)

# Fit model using best ridge alpha!
ridge_cv = ridge_cv.fit(Z_train,y_train)
```

```python
# Here is the optimal value of alpha
ridge_cv
```

```python
print(ridge_cv.score(Z_train,y_train))##TRAINING AND Testingsignificance level  should be below 5% , Here is not yet.
print(ridge_cv.score(Z_test,y_test))
```

```python
####Lasso implementation
# Imports similar to Ridge
from sklearn.linear_model import Lasso, LassoCV
```

```python
# Reminders
print(" OLS ".center(18, "="))
print(ols.score(Z_train, y_train))
print(ols.score(Z_test, y_test))
print()
print(" Ridge ".center(18, "="))
print(ridge_cv.score(Z_train, y_train))
print(ridge_cv.score(Z_test, y_test))
```

```python
# Set up a list of Lasso alphas to check.
l_alphas = np.logspace(-3,1,100)

# Cross-validate over our list of Lasso alphas.
lasso_cv = LassoCV (alphas=l_alphas, cv = 5, max_iter=500)

# Fit model using best ridge alpha!
lasso_cv.fit(Z_train,y_train)
```

```python
# Here is the optimal value of alpha
lasso_cv.alpha_
```

```python
print(" Ridge ".center(18, "="))
print(ridge_cv.score(Z_train, y_train))
print(ridge_cv.score(Z_test, y_test))
print(" Lasso ".center(18, "="))
print(lasso_cv.score(Z_train, y_train))### it is also better results.
print(lasso_cv.score(Z_test, y_test))  ##
```

#Train test split

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
X=df.drop("tip",axis=1)
y=df["tip"]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
X_train
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
y_train.shape
```

#Standardization (standard scaler)

```python
sc = StandardScaler()
Z_train= sc.fit_transform(X_train)
Z_test= sc.transform(X_test)
```

#linear regression

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression().fit(Z_train,y_train)
```

```python
lr.score(Z_train,y_train)#overfitting datset
```

```python
lr.score(Z_test,y_test)#overfitting dataset
```

#Lasso Regression(-L1-Regularization)

```python
from sklearn.linear_model import Lasso,LassoCV
```

```python
lasso_lr=Lasso(alpha=5,max_iter=500,tol=0.0001)
lasso_lr.fit(Z_train,y_train)
```

```python
lasso_lr.score(Z_train,y_train)
```

```python
lasso_lr.score(Z_test,y_test)
```

#Ridge-L2 Regularization

```python
from sklearn.linear_model import Ridge
```

```python
ridge_lr=Ridge(alpha=5,max_iter=500,tol=0.001)
ridge_lr.fit(Z_train,y_train )
```

```python
ridge_lr.score(Z_train,y_train)
```

```python
ridge_lr.score(Z_test,y_test)
```
