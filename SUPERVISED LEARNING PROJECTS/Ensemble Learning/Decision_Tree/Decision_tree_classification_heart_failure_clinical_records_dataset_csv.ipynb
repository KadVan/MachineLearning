```python
from sklearn.preprocessing import StandardScaler
standardscaler=StandardScaler()
columns_to_scale=["age","serum_sodium","platelets"]
df[columns_to_scale]=standardscaler.fit_transform(df[columns_to_scale])
```

```python
df.head()
```

```python
df=pd.get_dummies(df,columns=["platelets","serum_creatinine"],drop_first=True)
```

```python
df.head()
```

```python
X=df.drop(columns="DEATH_EVENT",axis=1)
y=df["DEATH_EVENT"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop(['DEATH_EVENT'], axis = 1),
                                                    df['DEATH_EVENT'],
                                                    test_size = 0.3,stratify=y,
                                                    random_state = 42)
```

```python
### Import model.
from sklearn.tree import DecisionTreeClassifier
```

```python
# Instantiate model with random_state = 42.
dt = DecisionTreeClassifier(random_state = 42)
```

```python

dt = DecisionTreeClassifier(
        max_depth = 5,
        min_samples_split = 7,
        min_samples_leaf = 3,
        ccp_alpha = 0.01
)
```

```python
# Fit model.
dt.fit(X_train,y_train)
```

```python

# Evaluate model.

print(f'Score on training set: {dt.score(X_train, y_train)}')
print(f'Score on testing set: {dt.score(X_test, y_test)}')
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(
    estimator = DecisionTreeClassifier(),
    param_grid = {
        'max_depth' : [2, 3, 5, 7],
        'min_samples_split' : [5, 10, 15, 20],
        'min_samples_leaf' : [2, 3, 4, 5, 6],
        'ccp_alpha' : [0, 0.001, 0.01, 0.1, 1, 10]
    },
    cv = 5,
    verbose = 1
)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_estimator_
```

```python
grid.best_score_
```

```python
# Evaluate model.
print(f'Score on training set: {grid.score(X_train, y_train)}')
print(f'Score on testing set: {grid.score(X_test, y_test)}')
```

```python
# Generate predictions on test set.
preds = grid.predict(X_test)
```

```python
# Import confusion_matrix.
from sklearn.metrics import confusion_matrix
```

```python
# Generate confusion matrix.
tn, fp, fn, tp = confusion_matrix(y_test,
                                  preds).ravel()

print(confusion_matrix(y_test,
                       preds))
```

```python
print('tn, fp, fn, tp',tn, fp, fn, tp)
```

```python
# Calculate sensitivity.

sens = tp / (tp + fn)

print(f'Sensitivity: {round(sens, 4)}')
```

```python
# Calculate specificity.

spec = tn / (tn + fp)

print(f'Specificity: {round(spec, 4)}')
```

Visualizing the Output of Decision Trees

```python
# Import plot_tree from sklearn.tree module.
from sklearn.tree import plot_tree

# Establish size of figure.
plt.figure(figsize = (50, 30))

# Plot our tree.
plot_tree(grid.best_estimator_,
          feature_names = X_train.columns,
          class_names = ['Dead', 'Survived'],
          filled = True);
```
