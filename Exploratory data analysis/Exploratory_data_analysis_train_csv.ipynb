```python
#no 1There are two columns that pertain to how many family members are on the boat for a given person. Create a new column called FamilyCount which will be the sum of those two columns.
```

```python
# Number 1:
# Creating new column named FamilyCount which combines siblings and parents for a family count
train['FamilyCount'] = (train['SibSp'] + train['Parch'])
```

```python
# Checking to see if new column worked
train.head()
```

```python
 #Number2---Reverends have a special title in their name. Create a column called IsReverend: 1 if they're a preacher, 0 if they're not.

# Creating a new column identifying if a passenger is a reverend
# Reverends have title "Rev." in their name. Setting new column to equal the "Name" column
train['isReverend'] = 0
```

```python
#Calling train to see if changes worked for IsReverend column
train.head()
```

```python
#Filtering through new column to change all names with "Rev" to 1
train.loc[train['Name'].str.contains('Rev'), 'isReverend'] = 1
```

```python
# Calling the train to see if the changes worked
train.tail()
```

```python
##no 3--In order to feed our training data into a classification algorithm, we need to convert our categories into 1's and 0's using pd.get_dummies
#Familiarize yourself with the pd.get_dummies documentation
#Create 3 columns: Embarked_C, Embarked_Q and Embarked_S. These columns will have 1's and 0's that correspond to the C, Q and S values in the Embarked column

```

```python
 #Using the pd.get_dummies
dum_embark = pd.get_dummies(train['Embarked'])

#Creating new column names that correspond to the C, Q and S values in the Embarked column
dum_embark.columns = ['Embarked_C', 'Embarked_Q', 'Embarked_S']

#calling the new dum_embark df to see if changes worked
dum_embark.head()
```

```python
#Concatinating titanc and dum_embark df so that they are one new df
new_train = pd.concat((train,dum_embark), axis =1)
#Printing the new df to make sure changes worked
new_train.head()
```

```python
#Do the same thing for Sex  column
```

```python
# Doing same thing we did with Embark columns for the Sex column
sex = pd.get_dummies(train['Sex'])
#Concatinating new_train df and sex df so that they are one new df
final_train = pd.concat((new_train, sex), axis =1)
final_train.head()
```
