#Time Resampling

```python
df_apple.head()
```

```python
df_apple.resample('A').mean()
```

```python
df_apple.resample('A').min()
```

```python
df_apple.resample('A').max()
```

```python
df_apple.resample(rule='A').max()['Open'].plot()
```

```python
df_apple.resample(rule='SM').max()['High'].plot()
```

```python
df_apple.resample(rule='QS').max()['High'].plot()
```

```python
df_apple['Open'].resample(rule='A').mean().plot(kind='bar')
```

```python
df_apple['Open'].resample(rule='M').max().plot(kind='bar',figsize=(15,8))
```
