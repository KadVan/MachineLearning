```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
import pandas as pd
```

```python
sns.set_style('whitegrid')
```

```python
#titanic = sns.load_dataset('titanic')
# sns.load_dataset('titanic')
titanic = pd.read_csv("/content/titanic.csv")
```

```python
titanic.head()
```

```python
sns.jointplot(x='fare',y='age',data=titanic)
```

```python
sns.distplot(titanic["fare"],kde=False)
```

```python
 #cretae boxlpot on class and age
 sns.boxplot(x='pclass',y='age',data=titanic)
```

```python
#create Swarmplot on class and age
sns.swarmplot(x='pclass',y='age',data=titanic)
```

```python
# Create count plot on sex
sns.countplot("sex",data=titanic)
```

```python
# for heat maps, indexing / correaltions needs to be established
titanic = titanic.corr()
sns.heatmap(titanic,cmap='coolwarm')
plt.title('titanic.corr()')
```

```python
g = sns.FacetGrid(data=titanic,col="age")
g.map(sns.distplot,"survived",kde=False)
```

```python
# initialize the FacetGrid object
g = sns.FacetGrid(data=titanic, col="age", row="pclass")

g.map(plt.hist, "fare")
g.add_legend()
```
